# Práctica de API request

```python
import os
from dotenv import load_dotenv
```

Se importan las librerias

```python
load_dotenv()

client_id = os.getenv("CLIENT_ID")
client_secret = os.getenv("CLIENT_SECRET")
```

Cargar credenciales

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
spotify = spotipy.Spotify(auth_manager=auth_manager)
```

Conexión con la API

```python
artist_id = "1uNFoZAHBGtllmzznpCI3s"

Justinbieber= 'spotify:artist:' + artist_id
results = spotify.artist_top_tracks(Justinbieber)
results["tracks"][0]
```

ID del artista para obtener el top 10

```python
results["tracks"][0]["name"]
```

Resultados:La salida corresponde a una lista de diccionarios con las claves: song,popularity y duration

```python
top_10 = []

for track in results['tracks'][:10]:
    top_10.append({"name": track["name"], "popularity": track["popularity"], "duration": track["duration_ms"]})

top_10
```
